### Assessing Data

```python
def assess_data(df, name):
    print(f"\nAssessing {name} Dataset:")
    print("\n1. Basic Information:")
    print(df.info())

    print("\n2. First few rows:")
    print(df.head())

    print("\n3. Missing Values:")
    print(df.isnull().sum())

    print("\n4. Duplicate Checks:")
    print(f"Full duplicates: {df.duplicated().sum()}")

    if 'order_id' in df.columns:
        print(f"Order ID duplicates: {df.duplicated('order_id').sum()}")
    if 'customer_id' in df.columns:
        print(f"Customer ID duplicates: {df.duplicated('customer_id').sum()}")

    print("\n5. Basic Statistics:")
    print(df.describe())

def check_invalid_values(orders_df, order_items_df, customers_df):
    print("\nChecking Invalid Values:")

    # 1. Check ID formats
    print("\n1. ID Format Checks:")
    # ID length check
    print("Order ID lengths:", orders_df['order_id'].str.len().value_counts())
    print("Customer ID lengths:", customers_df['customer_id'].str.len().value_counts())
    print("Product ID lengths:", order_items_df['product_id'].str.len().value_counts())

    # 2. Check Date Logic
    print("\n2. Date Logic Checks:")
    orders_df_dates = orders_df.copy()
    date_columns = ['order_purchase_timestamp', 'order_approved_at',
                   'order_delivered_carrier_date', 'order_delivered_customer_date',
                   'order_estimated_delivery_date']

    for col in date_columns:
        orders_df_dates[col] = pd.to_datetime(orders_df_dates[col])

    # Check if delivery date is before purchase date
    invalid_dates = orders_df_dates[
        orders_df_dates['order_delivered_customer_date'] <
        orders_df_dates['order_purchase_timestamp']
    ]
    print(f"Orders with delivery date before purchase: {len(invalid_dates)}")

    # 3. Check price and freight logic
    print("\n3. Price and Freight Logic:")
    print(f"Items with zero or negative price: {len(order_items_df[order_items_df['price'] <= 0])}")
    print(f"Items with negative freight: {len(order_items_df[order_items_df['freight_value'] < 0])}")

def check_inaccurate_values(orders_df, order_items_df):
    print("\nChecking Inaccurate Values:")

    # 1. Extreme price checks
    print("\n1. Price Outliers:")
    price_stats = order_items_df['price'].describe()
    q1 = price_stats['25%']
    q3 = price_stats['75%']
    iqr = q3 - q1
    price_outliers = order_items_df[
        (order_items_df['price'] < (q1 - 1.5 * iqr)) |
        (order_items_df['price'] > (q3 + 1.5 * iqr))
    ]
    print(f"Number of price outliers: {len(price_outliers)}")
    print(f"Price range of outliers: {price_outliers['price'].min()} to {price_outliers['price'].max()}")

    # 2. Delivery time checks
    orders_df_dates = orders_df.copy()
    for col in orders_df_dates.select_dtypes(include=['object']).columns:
        if 'date' in col or 'timestamp' in col:
            orders_df_dates[col] = pd.to_datetime(orders_df_dates[col])

    delivery_time = (orders_df_dates['order_delivered_customer_date'] -
                    orders_df_dates['order_purchase_timestamp']).dt.days

    print("\n2. Delivery Time Statistics:")
    print(delivery_time.describe())

def check_inconsistent_values(orders_df, customers_df):
    print("\nChecking Inconsistent Values:")

    # 1. Check status flow consistency
    print("\n1. Order Status Consistency:")
    print(orders_df['order_status'].value_counts())

    # 2. Check state codes consistency
    print("\n2. State Codes:")
    print(customers_df['customer_state'].value_counts())

    # 3. Check city names consistency
    print("\n3. City Names Consistency:")
    # Check for cities with multiple spellings
    city_state_groups = customers_df.groupby('customer_state')['customer_city'].unique()
    for state, cities in city_state_groups.items():
        # Look for similar city names using lowercase comparison
        city_lower = [city.lower() for city in cities]
        duplicates = set([x for x in city_lower if city_lower.count(x) > 1])
        if duplicates:
            print(f"\nPossible inconsistent city names in {state}:")
            for dup in duplicates:
                print(f"- {dup}")
```

```python
# Assess each dataset
for df, name in [(orders_df, "Orders"),
                 (order_items_df, "Order Items"),
                 (customers_df, "Customers")]:
    assess_data(df, name)

# Run the checks
check_invalid_values(orders_df, order_items_df, customers_df)
check_inaccurate_values(orders_df, order_items_df)
check_inconsistent_values(orders_df, customers_df)
```

### Cleaning Data

```python
def clean_data_for_rfm():
    # Copy dataframe
    orders_clean = orders_df.copy()
    order_items_clean = order_items_df.copy()

    print("Before Cleaning:")
    print("Total orders:", len(orders_df))
    print("Orders by status:\n", orders_df['order_status'].value_counts())
    print("\nMissing values in delivery date:", orders_df['order_delivered_customer_date'].isnull().sum())

    # 1. Konversi tanggal
    date_columns = ['order_purchase_timestamp', 'order_delivered_customer_date']
    for col in date_columns:
        orders_clean[col] = pd.to_datetime(orders_clean[col])

    # 2. Filter delivered orders
    orders_clean = orders_clean[
        (orders_clean['order_status'] == 'delivered') &
        (orders_clean['order_delivered_customer_date'].notna())
    ]

    # 3. Hitung monetary value
    order_values = order_items_clean.groupby('order_id')['price'].sum().reset_index()
    orders_clean = orders_clean.merge(order_values, on='order_id', how='left')

    # 4. TAMBAHAN: Cek distribusi orders per customer sebelum aggregasi
    customer_order_counts = orders_clean.groupby('customer_id')['order_id'].count().reset_index()
    print("\nDistribusi orders per customer sebelum aggregasi:")
    print(customer_order_counts['order_id'].value_counts())

    # 5. TAMBAHAN: Cek beberapa sample multiple orders
    print("\nSample multiple orders per customer:")
    print(orders_clean[orders_clean['customer_id'].isin(
        customer_order_counts[customer_order_counts['order_id'] > 1]['customer_id'].head()
    )][['customer_id', 'order_id', 'order_purchase_timestamp']])

    return orders_clean
```

```python
cleaned_orders = clean_data_for_rfm()
```

```python
# Explore karakteristik RFM customers
def calculate_rfm(cleaned_orders):
    current_date = cleaned_orders['order_delivered_customer_date'].max()

    # Pastikan aggregasi menghitung dengan benar
    customer_metrics = cleaned_orders.groupby('customer_id').agg({
        'order_delivered_customer_date': lambda x: (current_date - x.max()).days,  # Recency
        'order_id': 'nunique',  # Frequency - gunakan nunique untuk memastikan counting unique orders
        'price': 'sum'  # Monetary
    }).reset_index()

    customer_metrics.columns = ['customer_id', 'recency', 'frequency', 'monetary']

    # Print validasi
    print("\nValidasi frequency calculation:")
    print(customer_metrics['frequency'].value_counts().sort_index())

    return customer_metrics
```

```python
# Jalankan RFM calculation
rfm_df = calculate_rfm(cleaned_orders)

# Tampilkan summary statistik
print("\nRFM Metrics Summary:")
print("\nRecency (days):")
print(rfm_df['recency'].describe())

print("\nFrequency (count of orders):")
print(rfm_df['frequency'].describe())

print("\nMonetary (total spent):")
print(rfm_df['monetary'].describe())

# Tambahan explorasi
print("\nKorelasi antar metrics:")
print(rfm_df[['recency', 'frequency', 'monetary']].corr())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Siapa saja pelanggan yang berpotensi menjadi brand ambassador untuk produk kita?

```python
def analyze_potential_ambassadors(rfm_df):
    # Karena frequency = 1 untuk semua, fokus pada recency dan monetary
    rfm_df['R'] = pd.qcut(rfm_df['recency'], q=4, labels=['4','3','2','1'])  # 4 = recent
    rfm_df['M'] = pd.qcut(rfm_df['monetary'], q=4, labels=['1','2','3','4'])  # 4 = high value

    # Identifikasi potential brand ambassadors (recent buyers dengan high value)
    potential_ambassadors = rfm_df[
        (rfm_df['R'] == '4') &  # recent customers
        (rfm_df['M'] == '4')    # high monetary value
    ]

    print("\nPotential Brand Ambassadors Summary:")
    print(f"Jumlah kandidat: {len(potential_ambassadors)}")
    print("\nKarakteristik:")
    print(potential_ambassadors[['recency', 'monetary']].describe())

    # Visualisasi
    plt.figure(figsize=(10,6))
    plt.scatter(rfm_df['recency'], rfm_df['monetary'], alpha=0.5, color='gray')
    plt.scatter(potential_ambassadors['recency'], potential_ambassadors['monetary'],
                alpha=0.7, color='red', label='Potential Ambassadors')
    plt.xlabel('Recency (days)')
    plt.ylabel('Monetary Value (R$)')
    plt.title('Potential Brand Ambassadors Identification')
    plt.legend()
    plt.show()

    return potential_ambassadors
```

```python
ambassadors = analyze_potential_ambassadors(rfm_df)
```

### Pertanyaan 2: Pelanggan mana yang perlu diberikan program retensi khusus?

```python
def analyze_retention_targets(rfm_df):
    # Identifikasi pelanggan untuk retensi (high value tapi sudah lama tidak berbelanja)
    retention_targets = rfm_df[
        (rfm_df['R'] == '1') &  # not recent
        (rfm_df['M'] == '4')    # high value
    ]

    print("\nRetention Targets Summary:")
    print(f"Jumlah target: {len(retention_targets)}")
    print("\nKarakteristik:")
    print(retention_targets[['recency', 'monetary']].describe())

    # Visualisasi distribusi waktu sejak pembelian terakhir
    plt.figure(figsize=(10,6))
    plt.hist(retention_targets['recency'], bins=30)
    plt.xlabel('Days Since Last Purchase')
    plt.ylabel('Number of Customers')
    plt.title('Distribution of Time Since Last Purchase - Retention Targets')
    plt.show()

    return retention_targets
```
